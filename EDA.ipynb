```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_Mental_Stress_and_Coping_Mechanisms.csv")
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop("Student ID", axis=1, inplace= True)
df.head()
```

```python
bins = [0,3,6,10]
labels = ["Low", "Medium","High"]
df["Stress Level Category"] = pd.cut(df["Mental Stress Level"], bins = bins, labels = labels)
df.head()
```

```python
binary_cols = ["Counseling Attendance","Family Mental Health History","Medical Condition"]
for cols in binary_cols:
    df[cols] = df[cols].map({"Yes":1, "No":0})

df.head()
```

**Univariate Analysis**

```python
df.columns
```

```python
#Numerical Columns
num_cols = [
    "Age",
    "Academic Performance (GPA)",
    "Study Hours Per Week",
    "Social Media Usage (Hours per day)",
    "Sleep Duration (Hours per night)",
    "Physical Exercise (Hours per week)",
    "Family Support",
    "Financial Stress",
    "Peer Pressure",
    "Relationship Stress",
    "Mental Stress Level",
    "Diet Quality",
    "Cognitive Distortions",
    "Substance Use"
]

# Histograms for numeric columns
df[num_cols].hist(bins=15, figsize=(15, 12))
plt.tight_layout()
plt.show()
```

```python
#Categorical Columns
cat_cols = [
    "Gender",
    "Counseling Attendance",
    "Stress Coping Mechanisms",
    "Family Mental Health History",
    "Medical Condition"
]

for col in cat_cols:
    print(f"--- {col} ---")
    print(df[col].value_counts(), "\n")
```

**bivariate analysis**

```python
#bivariate analysis
continuous_features = [
    "Age", "Academic Performance (GPA)", "Study Hours Per Week",
    "Social Media Usage (Hours per day)", "Sleep Duration (Hours per night)",
    "Physical Exercise (Hours per week)", "Family Support", "Financial Stress",
    "Peer Pressure", "Relationship Stress", "Diet Quality",
    "Cognitive Distortions", "Substance Use"
]
target = "Mental Stress Level"

#Strip Plot
for feature in continuous_features:
    plt.figure(figsize=(8, 4))
    sns.stripplot(data=df, x=feature, y=target, jitter=True)
    plt.title(f"{feature} vs {target} (Strip Plot)")
    plt.xlabel(feature)
    plt.ylabel(target)
    plt.show()
```

```python
#box-plot
for feature in continuous_features:
    plt.figure(figsize=(8, 4))
    # Here, 'Stress Levels' is on the x-axis (categorical) and the continuous feature on the y-axis
    sns.boxplot(data=df, x=target, y=feature)
    plt.title(f"{feature} Distribution by {target}")
    plt.xlabel(target)
    plt.ylabel(feature)
    plt.show()
```

```python
# Box plots: Categorical features vs. Mental Stress Level
categorical_features = ["Counseling Attendance", "Stress Level Category",
                        "Family Mental Health History", "Medical Condition"]

for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x=feature, y=target)
    plt.title(f"{target} distribution by {feature}")
    plt.xlabel(feature)
    plt.ylabel(target)
    plt.show()
```

```python
# count plot: Example of two categorical variables
if "Gender" in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x="Gender", hue="Stress Level Category")
    plt.title("Distribution of Stress Categories by Gender")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.legend(title="StressCategory")
    plt.xticks(rotation=45)
    plt.show()
```

**Multivariate Analysis**

```python
# Multivariate Analysis
# Correlation Matrix: For continuous variables (including target)
plt.figure(figsize=(12, 10))
numeric_cols = continuous_features + [target]
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# Pair Plot: Visualize pairwise relationships for a subset of features
subset_features = ["Study Hours Per Week", "Sleep Duration (Hours per night)",
                   "Financial Stress", "Peer Pressure", target]
sns.pairplot(df[subset_features])
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# creating categorical stress level
def label_stress(row):
    if row <= 3:
        return 'Low'
    elif row <= 6:
        return 'Medium'
    else:
        return 'High'

df['Stress_Category'] = df['Mental Stress Level'].apply(label_stress)

# plotting the bar chart
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Stress_Category', order=['Low', 'Medium', 'High'], palette='Set2')
plt.title('Distribution of Stress Levels')
plt.xlabel('Stress Category')
plt.ylabel('Count')
plt.show()
```
